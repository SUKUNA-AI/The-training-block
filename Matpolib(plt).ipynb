```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,11.1,0.1)
y = np.sin(x)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Функция синуса от x')
plt.plot(x,y)
plt.show()
```

```python
fig = plt.figure()

# Добавляем набор осей
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (диапазон от 0 до 1)

# Наносим график на этот набор осей
axes.plot(x, y)

plt.show()
```
