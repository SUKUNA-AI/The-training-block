```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\ilyak\\Desktop\\Python\\Machine Learning Python\\DATA\\Advertising.csv')
df.head()
```

Полиноминальная регрессия

```python
X = df.drop('sales',axis=1)
y = df['sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_converter = PolynomialFeatures(degree=4,include_bias=False)
poly_features = polynomial_converter.fit_transform(X)
poly_features.shape
```

```python
%%time
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size = 0.3, random_state = 21)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
test_predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
MAE = mean_absolute_error(y_test, test_predictions)
RMSE = np.sqrt(mean_squared_error(y_test, test_predictions))
```
