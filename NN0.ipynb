```python
import json

dict_n = {}
with open('purchase_log.txt', encoding= 'utf8') as f:
    for i, line in enumerate(f):
        purchase = json.loads(line)
        key = purchase['user_id']
        value = purchase['category']
        dict_n[key] = value
        
with open('visit_log.csv', 'r', encoding= 'utf8') as f, open('funnel.csv', 'w', encoding= 'utf8') as f2:
    for i, line in enumerate(f):
        line = line.strip().split(',')
        if line[0] in dict_n.keys():
            line.append(dict_n[line[0]])
            ad_line = ','.join(line)
            
            f2.write(ad_line + '\n')
```

```python
import sys
from collections import deque

def main():
    scanner = int(sys.stdin.readline)
    n = scanner()

    processing_time = [0] * (n + 1)
    graph = [[] for _ in range(n + 1)]
    in_degree = [0] * (n + 1)

    for i in range(1, n + 1):
        line = scanner().split()
        t, = map(int, line[0].split(':'))
        processing_time[i] = t

        for dep in line[1:]:
            if dep:
                d, = map(int, dep.split('-')[0])
                graph[d].append(i)
                in_degree[i] += 1

    queue = deque([i for i in range(1, n + 1) if in_degree[i] == 0])

    finish_time = processing_time[:]
    while queue:
        process = queue.popleft()

        for neighbor in graph[process]:
            finish_time[neighbor] = max(finish_time[neighbor], finish_time[process] + processing_time[neighbor])
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0:
                queue.append(neighbor)

    minimum_completion_time = max(finish_time)
    print(minimum_completion_time)

if __name__ == "__main__":
    main()
```
